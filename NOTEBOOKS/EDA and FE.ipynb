```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
dataset.head()
```

```python
dataset.info()
```

## DATA CLEANING


```python
##check null values
dataset[dataset.isnull().any(axis=1)]
```

#  The dataset has two sets based on the regions from 122 index we can make new column based on the region 
# 1.Bejaja region 
# 2.sidi-bel-Abbes region
ADD a new column with the regionn


```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## remove the null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
##remove 122 
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
## fix spaces in columns using strip
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

## change the required columns as datatype

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

## changing other columns to data type


```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!="Classes":
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## lets save the dataset
df.to_csv("Algerain_forest_fires_cleaned_dataset.csv")
```
